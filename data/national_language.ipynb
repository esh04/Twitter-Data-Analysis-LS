```python
import pandas as pd
import numpy as np
import time
```

```python
from nltk.sentiment.vader import *
nltk.download('vader_lexicon')
```

```python
sentanalyzer=SentimentIntensityAnalyzer()

def sentiment(sentence):
  sent_dict=sentanalyzer.polarity_scores(sentence)

  if sent_dict['compound'] >=0.05 :
    return "positive"
  elif sent_dict['compound'] <= -0.05:
    return "negative"
  else:
    return "neutral"
```

```python
data=pd.read_json("/content/text-query-tweets2.json",lines=True)
data
```

```python
import goslate
import aiohttp
import asyncio
gs=goslate.Goslate()
#print(gs.translate('வணக்கம் உலகம்','en'))
```

```python


sentiment_list=[]

for i in data.index:
  sentiment_list.append(sentiment(str(data['content'][i])))
data['sentiments']=sentiment_list
```

```python

async def main():
  async with aiohttp.ClientSession() as session:
    tasks=[]
    translated=[]
    for i in data.index:
      sentiment_list.append(sentiment(str(data['content'][i])))
      if data['lang'][i] !='en':
        task = asyncio.ensure_future(gs.translate(str(data['content'][i]), 'en'))
      else:
        translated.append(str(data['content'][i]))
      translated.append(task)
      tasks.append(task)
    await asyncio(*tasks)
    data['content_translated']=translated
```
