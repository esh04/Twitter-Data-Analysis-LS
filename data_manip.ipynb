WORDCLOUDS

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd
import re

df = pd.read_csv("/content/drive/MyDrive/translated_data.csv")

eng=''
tamil=''
hindi=''
marathi=''
kannada=''
malayalam=''
telugu=''
stopwords=set(STOPWORDS)

for i in df.itertuples():
  if i.lang=='en':
    content=i.content
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    content=content.lower()
    eng=eng+' '+content
  if i.lang=='ta':
    content=i.translated
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    
    content=content.lower()
    tamil=tamil+' '+content
  if i.lang=='hi':
    content=i.translated
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    content=content.lower()
    hindi=hindi+' '+content
  if i.lang=='mr':
    content=i.translated
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    content=content.lower()
    marathi=marathi+' '+content
  if i.lang=='kn':
    content=i.translated
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    content=content.lower()
    kannada=kannada+' '+content
  if i.lang=='ml':
    content=i.translated
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    content=content.lower()
    malayalam=malayalam+' '+content
  if i.lang=='te':
    content=i.translated
    content=content.replace("#"," #")
    content=re.sub('[^A-Za-z# ]+','',content)
    content=" ".join(filter(lambda x:x[0]!='#', content.split()))
    content=content.lower()
    telugu=telugu+' '+content

    #print(i.translated)
  #print(i.content)

wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size =8).generate(eng)
 
# plot the WordCloud image                      
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
 
plt.show()
```
